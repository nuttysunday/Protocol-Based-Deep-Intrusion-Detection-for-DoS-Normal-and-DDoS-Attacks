```python
#imports
import pandas as pd
import numpy as np
```

```python
#merging all the csv's
df = pd.concat(
    map(pd.read_csv, 
        ["dataset/UNSW-NB15/a part of training and testing set/UNSW_NB15_training-set.csv",
         "dataset/UNSW-NB15/a part of training and testing set/UNSW_NB15_testing-set.csv"]), 
    ignore_index=True)
```

```python
#consider state sload dload later
df = df[['spkts','dpkts','sbytes','dbytes','dur','proto','attack_cat']]
df.rename(columns = {'attack_cat':'category'}, inplace = True)
```

```python
#we only take tcp and udp as by removing other columns we are only rejecting 16% of the data
df = df[(df['proto']=='udp') | (df['proto']=='tcp')]
```

```python
#replacing udp with 0 and tcp with 1
df['proto'] = np.where(df['proto'] == 'udp', 0, 1)
```

```python
df['category'].value_counts()
```

```python
#we only need  attack category of Normal type
df = df[df['category']=='Normal']
```
