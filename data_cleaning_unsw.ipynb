```python
#imports
import pandas as pd
import ipaddress
import numpy as np
```

```python
#reading csv
df = pd.read_csv("dataset/UNSW-NB15/UNSW-NB15_1.csv")
```

```python
df.head()
```

```python
#extracting column names from a features.csv and putting it as column header in the csv
df_name = pd.read_csv("dataset/UNSW-NB15/NUSW-NB15_features.csv",sep=",", encoding='cp1252')
columnn_name = df_name['Name']
df.columns = columnn_name
```

```python
df.head()
```

```python
#the attack_cat Nan needs to be replaced with Normal
df['attack_cat'].fillna('Normal', inplace=True)
```

```python
df['attack_cat'].value_counts()
```

```python
#taking only common features
df = df[['Spkts','Dpkts','sbytes','dbytes','dur','proto','Label','Stime','Ltime','sport','dsport','srcip','dstip','attack_cat']]
#replacing the column names with proper name
df.rename(columns = {'Spkts':'spkts','Dpkts':'dpkts','Label':'attack','Stime':'stime','Ltime':'ltime','sport':'sport','dsport':'dport','srcip':'saddr','dstip':'daddr','attack_cat':'category'}, inplace = True)
```

```python
df.head()
```

```python
#converting ip address into intgers
df['saddr'] =  df['saddr'].apply(lambda x:int(ipaddress.ip_address(x)))
df['daddr'] =  df['daddr'].apply(lambda x:int(ipaddress.ip_address(x)))
```

```python
df.head()
```

```python
df['proto'].value_counts()
```

```python
(494737+195478)/700000
```

```python
#we only take tcp and udp as by removing other columns we are only rejecting 2% of the data
df = df[(df['proto']=='udp') | (df['proto']=='tcp')]
```

```python
#replacing udp with 0 and tcp with 1
df['proto'] = np.where(df['proto'] == 'udp', 0, 1)
```

```python
df['category'].value_counts()
```

```python
#we only need  attack category of Normal type
df = df[df['category']=='Normal']
df.reset_index(drop=True,inplace=True)
```

```python
df.head()
```

```python
#splitting the dataset into train and saving the csv
df_train = df.iloc[:160000]
df_train.to_csv('dataset/cleaned_dataset/train/unsw_normal.csv',index=False)
```

```python
#splitting the dataset into test and saving the csv
df_test = df.iloc[663637:]
df_test.to_csv('dataset/cleaned_dataset/test/unsw_normal.csv',index=False)
```
